```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

### Las imágenes son matrices

Podemos representar numéricamente imágenes en formato de matrices. En el siguiente ejemplo vemos como, a raíz de una imágenes cargada como .jpg podemos representarla numéricamente. Ahora también vemos como, al dividir toda la matriz por un número x (10) y luego volver a representar la imágen, esta no cambia, y es porque, lo que es realmente importante es la correlación que existe entre un número y otro más que el número en si.

```python
plt.style.use('classic')

imagen = Image.open("./images/imagen_ejemplo_frida_bredesen.jpg")

plt.imshow(imagen)
```

```python
imagen_gr = imagen.convert('LA')
print(imagen_gr)
```

```python
imagen_matrix = np.array(list(imagen_gr.getdata(band=0)), float)

print(imagen_matrix)
```

```python
imagen_matrix.shape = (imagen_gr.size[1], imagen_gr.size[0])
print(imagen_matrix)
```

```python
print(imagen_matrix.shape)
```

```python
plt.imshow(imagen_matrix, cmap = 'gray')
plt.show()
```

```python
imagen_matrix_2 = imagen_matrix / 10 
print(imagen_matrix_2)
```

```python
plt.imshow(imagen_matrix_2, cmap='gray')
plt.show()
```

```python
# Viendo los máximos de cada matriz
print(np.max(imagen_matrix_2))
print(np.max(imagen_matrix))
```

```python
# Viendo los mínimos de cada matriz
print(np.min(imagen_matrix_2))
print(np.min(imagen_matrix))
```
