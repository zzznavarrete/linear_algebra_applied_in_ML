```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

### Descomposición SVD a una matriz de una imágen
La compresión de imágenes por SVD hace uso de que muy pocos de los valores singulares realmente representan información.
Los valores singulares se guardan de forma ordenada.

```python
plt.style.use('classic')
imagen = Image.open("./images/imagen_ejemplo_frida_bredesen.jpg")

plt.imshow(imagen)
```

```python
imagen_gr = imagen.convert('LA')

imagen_matrix = np.array(list(imagen_gr.getdata(band=0)), float)

imagen_matrix.shape = (imagen_gr.size[1], imagen_gr.size[0])
```

```python
U, D, V = np.linalg.svd(imagen_matrix)
```

```python
print(imagen_matrix.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

### Reconstruyendo nuestra imágen
Empezamos por 1 valor singular y vamos aumentando, podremos ver como poco a poco empieza a tomar forma nuestra imágen reconstruida en función a la cantidad de valores singulares utilizados para construirla.

```python
imagen_recons = np.matrix(U[:, :1]) * np.diag(D[:1]) * np.matrix(V[:1,:])

plt.imshow(imagen_recons, cmap='gray')
```

```python
i = 5

imagen_recons = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i,:])
plt.imshow(imagen_recons, cmap='gray')

titulo = "valores singulares = %s" % i

plt.title(titulo)
plt.show()
```
