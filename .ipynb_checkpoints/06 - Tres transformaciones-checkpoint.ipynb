```python
A = np.array([[3,7], [5,2]])

print(A)
```

```python
print('CÃ­rculo unitario: ')
graficarMatriz(np.array([[1,0], [0,1]]))

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()

print('CÃ­rculo unitario transformado: ')
graficarMatriz(A)

plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

**La matriz V rota el espacio.**

**La matriz D escala el espacio.**

**La matriz U rota de nuevo el espacio.**

La transformaciÃ³n del espacio de una matriz A es igual a la transformaciÃ³n de las matrices SVD (Valores singulares)

```python
U, D, V = np.linalg.svd(A)
```

```python
print(U)
```

```python
print('CÃ­rculo unitario: ')
graficarMatriz(np.array([[1,0], [0,1]]))

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()

# Aplicando la primer transformaciÃ³n (Matriz = V)
# Es una rotaciÃ³n

print('Primer rotaciÃ³n V: ')
graficarMatriz(V)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()

# Aplicando la primer transformaciÃ³n (Matriz = D)
# Es una 

print('Escala D: ')
graficarMatriz(np.diag(D).dot(V))

plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()


# Aplicando la segunda rotaciÃ³n (Matriz = U)
# Es una 

print('Seguna rotaciÃ³n U: ')
graficarMatriz(U.dot(np.diag(D).dot(V)))

plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```
