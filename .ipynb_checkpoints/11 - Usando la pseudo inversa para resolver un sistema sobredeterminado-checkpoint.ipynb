Un sistema de ecuaciones puede tener 0 soluciones, 1 solución o infinitas soluciones.
En el caso de que tuviera 1 solución, tenemos inversa. Ahora veremos como por medio de la pseudo inversa encontrar una solución x que tenga la sgte propiedad:

Si tenemos un sistema A * X = B. Usando la pseudo inversa (A_pse) queremos encontrar un X tal que la norma de A * X - B sea mínima.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)

# Definiendo 1ra ecuación
y_1 = -4*x+3

# 2da ecuación
y_2 = 2*x + 5

# 3ra ecuación
y_3 = -3*x + 1


plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-2, 2.5)
plt.ylim(-6, 6)

plt.show()
```

```python
matriz = np.array([[4, 1], [-2, 1], [3, 1]])
print(matriz)
```

```python
# Calculando la pseudo inversa
matriz_pse = np.linalg.pinv(matriz)
print(matriz_pse)
```

```python
# Vector solución (Estos valores vienen dado por el vector solución de cada ecuación)
b = np.array([[3], [5], [1]])
```

```python
resultado = matriz_pse.dot(b)
print(resultado)
```

```python

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-2, 2.5)
plt.ylim(-6, 6)

# Agregando el punto que acabamos de encontrar.
# La primer coordenada es el resultado 
plt.scatter(resultado[0], resultado[1])

plt.show()
```
