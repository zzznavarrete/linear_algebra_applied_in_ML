### Aplicando PCA a imágenes

Análisis de componentes principales o PCA, es una técnica muy útil para reducir la cantidad de dimensiones con las que estamos trabajando.

Muchas veces tenemos un conjunto de dimensiones muy grandes y lo que necesitamos es reducirlo para quedarnos, por ej, con el 80% de la información que contiene nuestro conjunto de datos.

Pero qué variables son las que necesitamos quedarnos para conservar esa información? 

Veamoslo aplicado a imágenes

```python
%matplotlib inline

import numpy as np
import imageio
import matplotlib.pyplot as plt
import pandas as pd
```

```python
im = imageio.imread("./images/s3/3.pgm")
im = im.astype(np.uint8)

print(im)
```

```python
# Evaluando el máximo de estos números para después normalizarla
print('Max original', end = ' ')
print(np.amax(im))
print('Min original', end = ' ')
print(np.amin(im))
```

```python
# Empezamos a normalizar

im_original = im

im = im / 255

print('Max original', end = ' ')
print(np.amax(im))
print('Min original', end = ' ')
print(np.amin(im))
```

```python
# Definimos que necesitamos 
fix, ax = plt.subplots(1, 2, figsize = (12,12),
                    subplot_kw = {'xticks' : [], 'yticks': []})

# Vemos la original
ax[0].imshow(im_original, cmap='gray')

# Vemos la normalizada, comprobamos que lo importante es la relación entre los números más que el número en si.
ax[1].imshow(im, cmap='gray')
```

```python
# Preparamos el entorno con la librería de imágenes que tenemos montadas
# Vemos que por la naturaleza de nuestro set de imágenes tenemos 10 fotos x cada individuo.

from glob import iglob

caras = pd.DataFrame([])

for path in iglob('./images/*/*.pgm'):
    im = imageio.imread(path)
    cara = pd.Series(im.flatten(), name=path)
    caras = caras.append(cara)
    
fig, axes = plt.subplots(5, 10, figsize = (15,8),
                    subplot_kw = {'xticks' : [], 'yticks': []},
                        gridspec_kw = dict(hspace = 0.01, wspace = 0.01))


for i, ax in enumerate(axes.flat):
    ax.imshow(caras.iloc[i].values.reshape(112, 92), cmap="gray")
```

```python
# Importamos PCA desde sklearn
from sklearn.decomposition import PCA
```

```python
# n_componentes necesarios para quedarme con el 50% de la variación de los datos.
caras_pca = PCA(n_components = 0.5)

# Hacemos un fit de las caras que tenemos cargadas.
# Y vemos que nos devuelve los parámetros que estamos utilizando
caras_pca.fit(caras)
```

```python
filas = 3
columnas = caras_pca.n_components_ // filas

fig, axes = plt.subplots(filas, columnas, figsize = (12,6),
                    subplot_kw = {'xticks' : [], 'yticks': []},
                        gridspec_kw = dict(hspace = 0.01, wspace = 0.01))


for i, ax in enumerate(axes.flat):
    ax.imshow(caras_pca.components_[i].reshape(112, 92), cmap="gray")
```

```python
print(caras_pca.n_components_)
```

```python
components = caras_pca.transform(caras)
proyeccion = caras_pca.inverse_transform(components)
```

```python
fig, axes = plt.subplots(5, 10, figsize = (15, 8),
                    subplot_kw = {'xticks' : [], 'yticks': []},
                        gridspec_kw = dict(hspace = 0.01, wspace = 0.01))


for i, ax in enumerate(axes.flat):
    ax.imshow(proyeccion[i].reshape(112, 92), cmap="gray")
```

```python
# n_componentes necesarios para quedarme con el 80% de la variación de los datos.
caras_pca = PCA(n_components = 0.8)

# Hacemos un fit de las caras que tenemos cargadas.
# Y vemos que nos devuelve los parámetros que estamos utilizando
caras_pca.fit(caras)
```

```python
# Aumenta de 6 components para quedarnos con el 50% de la info a 44 componentes
print(caras_pca.n_components_)
```

```python
# Volvemos a calcular 
components = caras_pca.transform(caras)
proyeccion = caras_pca.inverse_transform(components)
```

```python
# Vemos que la información que ahora se ve mucho más claro

fig, axes = plt.subplots(5, 10, figsize = (15, 8),
                    subplot_kw = {'xticks' : [], 'yticks': []},
                        gridspec_kw = dict(hspace = 0.01, wspace = 0.01))


for i, ax in enumerate(axes.flat):
    ax.imshow(proyeccion[i].reshape(112, 92), cmap="gray")
```

```python
# n_componentes necesarios para quedarme con el 99,9% de la variación de los datos.
caras_pca = PCA(n_components = 0.999)

# Hacemos un fit de las caras que tenemos cargadas.
# Y vemos que nos devuelve los parámetros que estamos utilizando
caras_pca.fit(caras)
```

```python
# Aumenta de 44 components para quedarnos con el 80% de la info a 388 componentes para quedarnos con 99.9%
print(caras_pca.n_components_)
```

```python
# Volvemos a calcular 
components = caras_pca.transform(caras)
proyeccion = caras_pca.inverse_transform(components)
```

```python
# Vemos que la información que ahora se ve mucho más claro incluso que con el 80%

fig, axes = plt.subplots(5, 10, figsize = (15, 8),
                    subplot_kw = {'xticks' : [], 'yticks': []},
                        gridspec_kw = dict(hspace = 0.01, wspace = 0.01))


for i, ax in enumerate(axes.flat):
    ax.imshow(proyeccion[i].reshape(112, 92), cmap="gray")
```
